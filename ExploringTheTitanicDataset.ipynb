# Building a Simple Classification Model

```python
# Import libraries

import sklearn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Exploring Titanic dataset

```python
# Loading data

titanic_df = pd.read_csv('datasets/titanic_train.csv')

titanic_df.head(10)
```

```python
# Data shape

titanic_df.shape
```

```python
# Drop irrelevant columns

titanic_df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], 'columns', inplace=True)

titanic_df.head()
```

```python
# Look for null values

titanic_df[titanic_df.isnull().any(axis=1)].count()
```

```python
# Drop row with missing values

titanic_df = titanic_df.dropna()
```

```python
# Data shape after dropped rows

titanic_df.shape
```

```python
# Checking nul values 

titanic_df[titanic_df.isnull().any(axis=1)].count()
```

```python
# Statistical overview of the data

titanic_df.describe()
```

## Visualizing Relationships in the Data

```python
titanic_df.head()
```

```python
# Plotting age vs survived to see the relationship

fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(titanic_df['Age'], titanic_df['Survived'])

plt.xlabel('Age')
plt.ylabel('Survived')
```

```python
# Plotting fare vs survived to see the relationship

fig, ax = plt.subplots(figsize=(12, 8))

plt.scatter(titanic_df['Fare'], titanic_df['Survived'])

plt.xlabel('Fare')
plt.ylabel('Survived')
```

```python
# Crosstab to who survived using sex (sex vs survived)

pd.crosstab(titanic_df['Sex'], titanic_df['Survived'])
```

```python
# Crosstab to who survived using pclass (pclass vs survived)

pd.crosstab(titanic_df['Pclass'], titanic_df['Survived'])
```

```python
# Checking correlation between observations

titanic_data_corr = titanic_df.corr()

titanic_data_corr
```

```python
# Plotting correlation using a heatmap

fig, ax = plt.subplots(figsize=(12, 10))

sns.heatmap(titanic_data_corr, annot=True)
```

```python
# Data preprocessing

from sklearn import preprocessing

label_encoding = preprocessing.LabelEncoder()
titanic_df['Sex'] = label_encoding.fit_transform(titanic_df['Sex'].astype(str))

titanic_df.head()
```

```python
# label encoding

label_encoding.classes_
```

```python
titanic_df = pd.get_dummies(titanic_df, columns=['Embarked'])

titanic_df.head()
```

```python
titanic_df = titanic_df.sample(frac=1).reset_index(drop=True)

titanic_df.head()
```

```python
# Save preprocessed data

titanic_df.to_csv('datasets/titanic_processed.csv', index=False)
```
